```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import sent_tokenize , word_tokenize
import glob
import re
import os
import numpy as np
import sys
import math 


def finding_all_unique_words_and_freq(words):
    words_unique = []
    word_freq = {}
    for word in words:
        if word not in words_unique:
            words_unique.append(word)
    for word in words_unique:
        word_freq[word] = words.count(word)
    return word_freq
def finding_freq_of_word_in_doc(word,words):
    freq = words.count(word)
        
def remove_special_characters(text):
    regex = re.compile('[^a-zA-Z0-9\s]')
    text_returned = re.sub(regex,'',text)
    return text_returned

def term_frequency(word,dict1):
    
    if word in dict1.keys():
        count_of_words=dict1.get(word)
        values = dict1.values()
        
        total=sum(values)
        return count_of_words/total
    else: 
        return 0
        
def inverse_term_frequency(z):
    len1=len(z)
    
    total=sum(z)
    return math.log10(len1/total)
    

all_words = []
dict_global={}
file_folder = 'C:\\Users\\SR\\IR\\*'
idx = 0
freq_dict={}
files_with_index = {}

Stopwords = set(stopwords.words('english'))

for file in glob.glob(file_folder):
    
    dict_local = {}
    fname = file
    file = open(file , "r")
    text = file.read()
    text = remove_special_characters(text)
    text = re.sub(re.compile('\d'),'',text)
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    
    words = [word for word in words if len(words)>0]
    
    words = [word.lower() for word in words]
    
   
    dict_local.update(finding_all_unique_words_and_freq(words))
    freq_dict[idx]=finding_all_unique_words_and_freq(words)
    
    files_with_index[idx] = os.path.basename(fname)
    idx = idx + 1
    dict_global.update(dict_local)
    
    
#print(files_with_index)   
#print(freq_dict)
count=1
doc_list=list(files_with_index.values())
print("Document in Corpus:- ")
for i in doc_list:
    print("Doc:-",count,i)
    count+=1


unique_words_all = set(dict_global.keys())
class Node:
    def __init__(self ,docId, freq = None):
        self.freq = freq
        self.doc = docId
        self.nextval = None
    
class SlinkedList:
    def __init__(self ,head = None):
        self.head = head


    
linked_list_data = {}
for word in unique_words_all:
    linked_list_data[word] = SlinkedList()
    linked_list_data[word].head = Node(1,Node)
word_freq_in_doc = {}
#print(linked_list_data)
idx = 1
freq_dict={}

for file in glob.glob(file_folder):
    file = open(file, "r")
    text = file.read()
    text = remove_special_characters(text)
    text = re.sub(re.compile('\d'),'',text)
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    words = [word for word in words if len(words)>0]
    words = [word for word in words if word not in Stopwords]
    words = [word.lower() for word in words]
    
    word_freq_in_doc = finding_all_unique_words_and_freq(words)
    freq_dict[idx]=word_freq_in_doc
#    print(word_freq_in_doc)
    for word in word_freq_in_doc.keys():
        linked_list = linked_list_data[word].head
        while linked_list.nextval is not None:
            linked_list = linked_list.nextval
        linked_list.nextval = Node(idx ,word_freq_in_doc[word])
    idx = idx + 1
#print(linked_list_data)
#print(word_freq_in_doc)
#print(freq_dict)

freq_list=list(freq_dict.values())
count=1
print("Term frequency of each term in every document:-")
for it in freq_list:
    print("Doc:-",count,it)
    count+=1

query = input('Enter your query:')
query = word_tokenize(query)

boolean_connector = []
cnt = 1
different_words = []

for word in query:
    if word.lower() != "and" and word.lower() != "or" and word.lower() != "not":
        different_words.append(word.lower())
    else:
        boolean_connector.append(word.lower())
#print(boolean_connector)
freq_list_of_all_words=[]
freq_list=[]

different_words=[word for word in different_words if word not in Stopwords]

total_files = len(files_with_index)
for word in different_words:
    idx = 0
    freq_list=[0]*total_files
    for file in glob.glob(file_folder):
        dict_local = {}
        fname = file
        file = open(file , "r")
        text = file.read()
        text = remove_special_characters(text)
        text = re.sub(re.compile('\d'),'',text)
        sentences = sent_tokenize(text)
        words = word_tokenize(text)
    
        words = [word for word in words if len(words)>0]
    
        words = [word.lower() for word in words]
        
   
        dict_local.update(finding_all_unique_words_and_freq(words))
        freq_dict[idx]=finding_all_unique_words_and_freq(words)
        freq_list[idx]=term_frequency(word,freq_dict[idx])
        files_with_index[idx] = os.path.basename(fname)
        idx = idx + 1
       
        
    freq_list_of_all_words.append(freq_list)
#print(freq_list_of_all_words)
for w in boolean_connector:
    if w=="not":
        boolean_connector.remove("not")    
empty=[]
zeroes_and_ones = []
zeroes_and_ones_of_all_words = []
idf=[]
for word in (different_words):
    if word.lower() in unique_words_all:
        zeroes_and_ones = [0] * total_files
        
        linkedlist = linked_list_data[word].head
        print(word)
        while linkedlist.nextval is not None:
            zeroes_and_ones[linkedlist.nextval.doc - 1] = 1
            linkedlist = linkedlist.nextval
           
        for i in range(len(query)-1):
            if query[i]== "not" and query[i+1]==word:
                
                for j in range(len(zeroes_and_ones)):
                    if zeroes_and_ones[j]==0:
                        zeroes_and_ones[j]=1;
                    elif zeroes_and_ones[j]==1:
                        zeroes_and_ones[j]=0;
        idf_term=inverse_term_frequency(zeroes_and_ones)       
        idf.append(idf_term)
       
        zeroes_and_ones_of_all_words.append(zeroes_and_ones) 
    else:
        zeroes_and_ones = [0] * total_files
        for i in range(len(query)-1):
            if query[i]== "not" and query[i+1]==word:
                
                for j in range(len(zeroes_and_ones)):
                    if zeroes_and_ones[j]==0:
                        zeroes_and_ones[j]=1;
                    elif zeroes_and_ones[j]==1:
                        zeroes_and_ones[j]=0;
                zeroes_and_ones_of_all_words.append(zeroes_and_ones)
         
        zeroes_and_ones_of_all_words.append(zeroes_and_ones)
        
        


#print(idf) 


for word in boolean_connector:
    word_list1 = zeroes_and_ones_of_all_words[0]
    word_list2 = zeroes_and_ones_of_all_words[1]
    if word == "and":
        bitwise_op = [w1 & w2 for (w1,w2) in zip(word_list1,word_list2)]
        zeroes_and_ones_of_all_words.remove(word_list1)
        zeroes_and_ones_of_all_words.remove(word_list2)
        zeroes_and_ones_of_all_words.insert(0, bitwise_op);
        
    elif word == "or":
        bitwise_op = [w1 | w2 for (w1,w2) in zip(word_list1,word_list2)]
        zeroes_and_ones_of_all_words.remove(word_list1)
        zeroes_and_ones_of_all_words.remove(word_list2)
        zeroes_and_ones_of_all_words.insert(0, bitwise_op);
      
        
files = []    
#print(zeroes_and_ones_of_all_words)
score=[]

if len(boolean_connector)==0:
    if len(different_words)==1:
        lis = zeroes_and_ones_of_all_words[0]
        cnt = 0
        for index in lis:
            if index == 1:
                files.append(files_with_index[cnt])
            cnt = cnt+1
        if len(files)==0:
            print("not found")
        else:
            print(files)
            
    elif len(different_words)>1:
        for i in range(len(idf)):
            x=idf[i]
            j=freq_list_of_all_words[i]
    
            temp_1=[]
            temp_1= [element * x for element in j]
            score.append(temp_1)
            
        

        final_score=[0]*total_files

        for j in range(len(score)):
            temp_2=zip(final_score,score[j])
            final_score=[x + y for (x, y) in temp_2]
    
        #print(final_score)
#final_score=final_score    
        rr_list=[]
        #print(files_with_index)
        file_list=list(files_with_index.values())
        while len(file_list) >0:
            curr_max=max(final_score)
            curr_max_index = final_score.index(curr_max)
            if curr_max>0:
                rr_list.append(file_list[curr_max_index])
#print(rr_list)
            file_list.remove(file_list[curr_max_index])
            final_score.remove(curr_max)
        
        if len(score)==0:
            print("Not Found")
        else:
            print(rr_list)
else:
    if len(boolean_connector)==(len(different_words)-1):
        lis = zeroes_and_ones_of_all_words[0]
        cnt = 0
        for index in lis:
            if index == 1:
                files.append(files_with_index[cnt])
            cnt = cnt+1
        if len(files)==0:
            print("not found")
        else:
            print(files)
    else:
        for i in range(len(idf)):
            x=idf[i]
            j=freq_list_of_all_words[i]
            temp_1=[]
            temp_1= [element * x for element in j]
            score.append(temp_1)
        #print(score)
        final_score=[0]*total_files
        for j in range(len(score)):
            temp_2=zip(final_score,score[j])
            final_score=[x + y for (x, y) in temp_2]
        #print(final_score)
#final_score=final_score    
        rr_list=[]
        #print(files_with_index)
        file_list=list(files_with_index.values())
        while len(file_list) >0:
            curr_max=max(final_score)
            curr_max_index = final_score.index(curr_max)
            if curr_max>0:
                rr_list.append(file_list[curr_max_index])
            file_list.remove(file_list[curr_max_index])
            final_score.remove(curr_max)
        if len(score)==0:
            print("Not Found")
        else:
            print(rr_list)
    
    
```
